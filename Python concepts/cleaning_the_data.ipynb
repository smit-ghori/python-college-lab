```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Name": [
        "Alice", "Bob", "Charlie", "David", "Eva",
        "Frank", "Grace", "Jack", "Jack", "Jack"
    ],
    "Age": [
        "25", "abc", np.nan, "30", "Unknown",
        "45", "twenty", "18", "", "60"
    ],
    "City": [
        "Delhi", "Mumbai", "Pune", "Ahmedabad", "Surat",
        "Rajkot", "Vadodara", "Jaipur", "Indore", "Bhopal"
    ]
})
```

```python
df
```

```python
df.dropna()
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df.dropna(subset=["Age"], inplace=True)  # returning new dataframe
df["Age"] = df["Age"].astype(int)
```

```python
df
```

```python
df.duplicated(subset=["Name"])
df.drop_duplicates(subset=["Name"], inplace=True)
```
